```python
import pandas as pd
import os
```

```python
path = '/content/drive/MyDrive/UnMazer Assignment'
```

```python
listings = pd.read_csv( os.path.join(path, 'listings.csv') )
neighbourhoods = pd.read_csv( os.path.join(path, 'neighbourhoods.csv') )
```

```python
print( listings )
print( neighbourhoods )
```

```python
#Replacing the values 'Private room', 'Shared room', 'Hotel room' with value 'Single room' for simplicity
listings['room_type'] = listings['room_type'].replace(['Private room','Hotel room','Shared room'],'Single room')
```

```python
filter1 = listings["room_type"]=="Entire home/apt"
filter2 = listings["room_type"]=="Single room"
```

```python
##  INSIGHT :  Multi room listings are much more popular, indicated by significantly higher reviews and number of reviews and much more
##             in-demand, indicated by significantly higher price and availability per year than single room listings. This means people 
##             prefer to travel to the place in groups rather than alone, indicating that the place might be a popular tourist destination.
```

```python
multiroom_avg_reviews_num = listings.where(filter1)['number_of_reviews'].mean()
singleroom_avg_reviews_num = listings.where(filter2)['number_of_reviews'].mean()

print(multiroom_avg_reviews_num)
print(singleroom_avg_reviews_num)
#Multi room listings have double number of average reviews
```

```python
multiroom_avg_reviews_pm = listings.where(filter1)['reviews_per_month'].mean()
singleroom_avg_reviews_pm = listings.where(filter2)['reviews_per_month'].mean()

print(multiroom_avg_reviews_pm)
print(singleroom_avg_reviews_pm)
#Multi room listings have 1.75 times number of reviews pm
```

```python
multiroom_avg_avai = listings.where(filter1)['availability_365'].mean()
singleroom_avg_avai = listings.where(filter2)['availability_365'].mean()

print(multiroom_avg_avai)
print(singleroom_avg_avai)
#Multi room listings have 1.35 times number of availability pm
```

```python
multiroom_avg_price = listings.where(filter1)['price'].mean()
singleroom_avg_price = listings.where(filter2)['price'].mean()

print(multiroom_avg_price)
print(singleroom_avg_price)
#Multi room listings have 2.65 times price
```
